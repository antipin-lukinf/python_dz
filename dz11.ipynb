```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train (1).csv') #считывание файла
```

```python
df
```

```python
df.query("Survived == 1")
```

выжившие пассажиры

```python
plt.plot(df['Age'])
```

Возраст пассажиров

```python
df.Pclass.value_counts() # кол-во строк по каждому району (столбец borough)
```

Количество пассажиров в каждом классе

```python
df.groupby('Sex', as_index=False).agg({'Age':'mean'}).sort_values('Age', ascending=True).round(2)
```

Средний возраст мужчин и женщин

```python
df.groupby(['Sex', 'Survived'], as_index=False)\
    .agg({'Age':'mean'})\
    .sort_values('Survived', ascending=True)\
    .round(2)
```

Средний возраст не выживших и выживших мужчин и женщин

```python
#plt.plot(df['Survived', 'Age']) 
df.sort_values('Fare').sort_values('PassengerId').plot(x="PassengerId", y=["Fare"])
plt.show()
```

Сколько потратил денег каждый из пассажиров

```python
df.groupby(['Pclass'], as_index=False)\
    .agg({'Survived':'sum'})\
    .sort_values('Pclass', ascending=True)\
    .round(2)
```
